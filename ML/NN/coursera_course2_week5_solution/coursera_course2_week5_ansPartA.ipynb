### IMPORTING REQUIRED LIBRARIES

```python
from keras.layers import Dense
from keras import Sequential
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

### STORING THE DATA

```python
data = pd.read_csv("concrete_data.csv")
data.head()
```

### DECLARING THE PREDICTORS AND TARGETS

```python
predictors = data.drop(['Strength'], axis = 1 )
targets = data['Strength']
```

```python
mse_50_times = []
```

### CREATING A FUNCTION TO CREATE A NN AND PREDICT THE VALUES

```python
def create_model_predict():
    x_train, x_test, y_train, y_test = train_test_split(predictors, targets, test_size=0.3 )
    nn_model = Sequential()
    nn_model.add(Dense(10, activation='relu', input_shape=(8,)))
    nn_model.add(Dense(10, activation='relu'))
    nn_model.add(Dense(1))
    nn_model.compile(optimizer='adam', loss='mean_squared_error')
    nn_model.fit(x_train, y_train, epochs=50, verbose=2, validation_split=0.1)
    yhat = nn_model.predict(x_test)
    #FINDING THE MEAN AND APPENDING IT TO A LIST
    mse_50_times.append(mean_squared_error(y_test, yhat))
```

```python
for i in range(50):
    create_model_predict()
```

### FINDING THE MEAN OF THE MEAN_SQUARED_ERRORS

```python
mse_50_times = np.asarray(mse_50_times)
mse_50_times.mean()
```

### FINDING THE STANDARD DEVIATION OF THE MEAN_SQUARED_ERRORS

```python
mse_50_times.std()
```
