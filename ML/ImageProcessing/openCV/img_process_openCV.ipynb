```python
from PIL import Image
import cv2
```

```python
def concat_image(img1, img2):
    combined = Image.new('RGB', size=(img1.width + img2.width , img1.height))
    combined.paste(img1, (0, 0))
    combined.paste(img2, (img1.width, 0))
    return combined
```
