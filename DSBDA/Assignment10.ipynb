```python
import pandas as pd
import seaborn as sns
sns.set()
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('iris.csv')
data.head()
```

```python
sns.histplot(data = data,  x=data['sepal_length'], hue = data['species'])
```

```python
sns.histplot(data = data,  x=data['sepal_width'], hue = data['species'])
```

```python
sns.histplot(data = data,  x=data['petal_length'], hue = data['species'])
```

```python
sns.histplot(data = data,  x=data['petal_width'], hue = data['species'])
```

```python
sns.boxplot( x = 'species',y='sepal_length', data = data)
```

```python
sns.boxplot( x = 'species',y='sepal_width', data = data)
```

```python
sns.boxplot( x = 'species',y='petal_length', data = data)
```

```python
sns.boxplot( x = 'species',y='petal_width', data = data)
```
